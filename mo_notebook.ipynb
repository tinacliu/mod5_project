```python
# to reduce the number of dimensions, try use PCA on highly correlated features
# limit_balance and bill_amount are highly correlated
from sklearn.preprocessing import StandardScaler
cols = ['limit_bal', 'bill_amt1', 'bill_amt2', 'bill_amt3', 'bill_amt4', 'bill_amt5', 'bill_amt6', 
        'months_late_1', 'months_late_2', 'months_late_3', 'months_late_4',
        'months_late_5', 'months_late_6']
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)

from sklearn.decomposition import PCA
fit = PCA()
pca_data = fit.fit_transform(X_scaled[cols])
```

```python
fit.explained_variance_ratio_
```

```python
# with 4 components we can explain over 80% of the variance in the data in our correlated columns
# turn this to a df and concatenate with other features
pca_df = pd.DataFrame(pca_data[:,:4], columns=['c1','c2','c3','c4'], index= X_scaled.index)
other_features = X_scaled.drop(cols, axis = 1)
X = pd.concat([pca_df, other_features], axis= 1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_curve, auc, accuracy_score, recall_score, precision_score, f1_score
clf_1 = KNeighborsClassifier()
clf_1.fit(X_train, y_train)
y_pred_proba = clf_1.predict_proba(X_test)[:,1]
y_pred = clf_1.predict(X_test)
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
print(auc(fpr,tpr))
print(recall_score(y_test, y_pred))
print (f1_score(y_test, y_pred))
```

```python
# education: group 4,5,6 into one bin as these are all unkowns

df['education'] = df['education'].map(lambda x: 4 if x >= 4 else x)
df.education.value_counts()
```

```python
df.rename({'pay_0':'pay_1'}, axis= 1, inplace= True)
```

```python
# pay_i: group 3 or more arrears into a new series
for i in range(1,7):
    df[f'pay_{i}'] = df[f'pay_{i}'].map(lambda x: 3 if x >= 3 else x)
```

```python
for i in range(1,7):    
    df[f'months_late_{i}'] = df[f'pay_{i}'].map(lambda x: x if x > 0 else 0)
    df[f'pay_{i}'] = df[f'pay_{i}'].map(lambda x: 1 if x <=0 else 0)
```

```python
df['sex'] = df['sex'].map({1:1, 2:0})
df.head(20)
```

```python
# calculate payment as a % of prev months balance
df_1 = df.copy()
for i in range(1,6):
    df_1[f'pay%_{i}'] = df_1[f'pay_amt{i}'] / df_1[f'bill_amt{i+1}']
```

```python
df_1 = df_1.dropna(axis=0)
```

```python
df_1.info()
```

```python
import numpy as np
df_1.replace([np.inf, -np.inf], np.nan,inplace=True)
df_1.dropna(axis= 0, inplace = True)
# df_1 has some feature engineering, df is just cleaned version of data
```

```python
X = df_1.drop('default', axis= 1)
y = df_1['default']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
clf_tree = DecisionTreeClassifier()
param_grid= {'max_depth': list(range(1,15)), 'min_samples_leaf': [10,100,1000]}
opt_tree = GridSearchCV(clf_tree, param_grid, scoring = 'f1', cv= 5)
opt_tree.fit(X_train, y_train)
best_tree = opt_tree.best_estimator_
print(opt_tree.best_params_)
```

```python
# best results using a max_depth of 3 and min samples leaf of 10
# explore other values of min sample leaf between 10 and 100
```

```python
clf_tree = DecisionTreeClassifier()
param_grid= {'max_depth': list(range(1,10)), 'min_samples_leaf': [10,20,30,40,50]}
opt_tree = GridSearchCV(clf_tree, param_grid, scoring = 'f1', cv= 5)
opt_tree.fit(X_train, y_train)
best_tree = opt_tree.best_estimator_
print(opt_tree.best_params_)
```

```python
pd.DataFrame(opt_tree.cv_results_)
```

```python
# a model with a max_depth of 3 and leaf sizes of 40 is not very complex so we choose a model
# with a similar score with higher complexity {max_depth: 6, min_samples_leaf: 20}
```

```python
from sklearn.metrics import roc_curve, auc, accuracy_score, recall_score, precision_score, f1_score
best_tree = DecisionTreeClassifier(max_depth = 6, min_samples_leaf= 20)
best_tree.fit(X_train, y_train)
y_pred = best_tree.predict(X_test)
y_pred_proba = best_tree.predict_proba(X_test)[:,1]
fpr,tpr, thresholds = roc_curve(y_test, y_pred_proba)
print(auc(fpr,tpr))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

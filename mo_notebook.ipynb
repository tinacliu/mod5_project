```python
import pandas as pd

df = pd.read_csv('credit_card_data.csv', skiprows = 1)

df.set_index('ID', inplace=True)

df.head()

```

```python
df.columns = df.columns.str.lower()
df.rename({'default payment next month': 'default'}, axis= 1, inplace= True)
```

```python
# calculate payment as a % of prev months balance
for i in range(1,6):
    df[f'pay%_{i}'] = df[f'pay_amt{i}'] / df[f'bill_amt{i+1}']
```

```python
df.head()
```

```python
# education: group 4,5,6 into one bin as these are all unkowns

df['education'] = df['education'].map(lambda x: 4 if x >= 4 else x)
df.education.value_counts()
```

```python
df['arrears_1'] = df['pay_0'].map(lambda x: 3 if x >= 3 else x)

# pay_i: group 3 or more arrears into a new series
for i in range(2,7):
    df[f'arrears_{i}'] = df[f'pay_{i}'].map(lambda x: 3 if x >= 3 else x)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df_1 = df.dropna(axis=1)
df_1.info()
```

```python
df_1['sex'] = df_1['sex'].map({1:1, 2:0})
df_1.head()
```

```python
df_1.drop(['pay_0','pay_2','pay_3','pay_4','pay_5','pay_6'], axis= 1, inplace= True)
df_1.info()
```

```python
cat_cols = ['education', 'marriage', 'arrears_1', 'arrears_2','arrears_3','arrears_4','arrears_5','arrears_6']
df_1 = pd.get_dummies(df_1, columns= cat_cols, drop_first= True)
df_1
```

```python
df_1.info()
```

```python
X = df_1.drop('default', axis= 1)
y = df_1['default']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.decomposition import PCA
fit = PCA()
pca_train = fit.fit_transform(X_train_scaled)
pca_train
```

```python
print (fit.explained_variance_ratio_[:20].sum())
```

```python
fit.components_[:20]
```

```python
pca_test = fit.transform(X_test_scaled)
pca_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_curve, auc, accuracy_score, recall_score, precision_score, f1_score
clf_1 = KNeighborsClassifier()
clf_1.fit(X_train_scaled,y_train)
y_hat_test = clf_1.predict(X_test_scaled)
fpr, tpr, thresholds = roc_curve(y_test, y_hat_test)
print(auc(fpr,tpr))
print(f1_score(y_test, y_hat_test))
```

```python
clf_2 = KNeighborsClassifier()
clf_2.fit(pca_train, y_train)
y_hat_test = clf_2.predict(pca_test)
fpr, tpr, thresholds = roc_curve(y_test, y_hat_test)
print(auc(fpr,tpr))
print(f1_score(y_test, y_hat_test))
```

```python
# using PCA did not improve our AUC by much
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors': list(range(5,26,2))}
opt_model = GridSearchCV(clf_1, param_grid)
opt_model.fit(X_train_scaled, y_train)
best_clf_1 = opt_model.best_estimator_
opt_model.best_params_
```

```python
opt_model_2 = GridSearchCV(clf_2, param_grid)
opt_model_2.fit(pca_train, y_train)
best_clf_2 = opt_model_2.best_estimator_
opt_model_2.best_params_
```

```python
# once again the difference in the PCA model and the regular model is very small
y_hat_test = best_clf_1.predict(X_test_scaled)
fpr, tpr, thresholds = roc_curve(y_test, y_hat_test)
print(auc(fpr,tpr))
```

```python
from sklearn.tree import DecisionTreeClassifier
clf_tree = DecisionTreeClassifier()
param_grid= {'max_depth': list(range(3,10)), 'min_samples_leaf': list(range(5,26,5))}
opt_tree = GridSearchCV(clf_tree, param_grid, scoring = 'roc_auc', cv= 5)
opt_tree.fit(X_train, y_train)
best_tree = opt_tree.best_estimator_
print(opt_tree.best_params_)
```
